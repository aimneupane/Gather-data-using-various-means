```python
headers = {
    "User-Agent": "Mozilla/5.0"
}
jobs = []
company_names = []

# change range as needed (example: first 10 pages)
for page in range(1, 4):
    url = f"https://www.jobsnepal.com/top-job?page={page}"
    res = requests.get(url, headers=headers)
    soup = BeautifulSoup(res.text, "html.parser")

    company = soup.find_all('div', class_='card-inner')

    # stop if no data (important!)
    if not company:
        break

    for i in company:
        job = i.find('h2')
        name = i.find('p')

        jobs.append(job.text.strip() if job else None)
        company_names.append(name.text.strip() if name else None)

# create dataframe
df = pd.DataFrame({
    'job': jobs,
    'company_name': company_names
})
df
```
